## 1. Introducción

La salud mental es un componente crucial del bienestar general de una población. Sin embargo, ha sido un tema frecuentemente marginado tanto en discusiones públicas como en políticas de salud. 

En los últimos años, y más notablemente debido a los impactos de la pandemia de COVID-19, la importancia de entender y abordar la salud mental ha llegado al primer plano de la conciencia pública. En este contexto, el presente proyecto busca explorar la situación de la salud mental en Lisboa y la región de Vale do Tejo, en Portugal, a través de un análisis de datos.

Este estudio se centra en el número de consultas psicológicas realizadas en diversas instituciones en los últimos cinco años, desde 2018 hasta 2022. Utilizando técnicas de análisis de datos, el proyecto busca entender las tendencias y patrones en la demanda de servicios de salud mental, así como explorar posibles correlaciones con otros indicadores, como el índice de felicidad en la región y en Europa.

Además, se plantean varias hipótesis relacionadas con la prevalencia de enfermedades mentales, la saturación de los sistemas de salud, y el impacto de factores estacionales y eventos globales en la salud mental. El objetivo es ofrecer una visión integral que pueda ser útil para futuras investigaciones, políticas de salud pública y estrategias de intervención.

## Hipótesis para el Proyecto de Salud Mental y Bienestar en Portugal
### Hipótesis Principales:

1. Deterioro de la Salud Mental en la Sociedad Moderna:
- La creciente necesidad de consultas psicológicas sugiere un deterioro general de la salud mental en la sociedad moderna.

2. Saturación de los Sistemas de Salud en Portugal:
- El incremento en las consultas psicológicas podría estar saturando los sistemas de salud nacionales en Portugal, evidenciando la necesidad de más profesionales en el campo de la salud mental.

3. Impacto de la COVID-19 en la Salud Mental:
- La pandemia de COVID-19 ha tenido un impacto significativo en la salud mental de la población, lo cual se refleja en un aumento de las consultas psicológicas durante los años de la pandemia.

### Hipótesis Adicionales:

1. Relación entre Felicidad Nacional e Incidencia de Consultas Psicológicas:
- ¿Existirá una correlación entre los indices de felicidad nacionales y la cantidad de consultas?

2. Factores Estacionales en la Salud Mental:
- Existen períodos del año donde las consultas psicológicas aumentan, posiblemente relacionados con factores como el clima, épocas festivas o periodos escolares.

```python
df_felicidad = pd.read_csv('../data/Data_limpia/Felicidad/df_felicidad.csv')
df_felicidad.drop(columns='Unnamed: 0', axis=1, inplace=True)
```

En este fragmento de código, se cargan dos conjuntos de datos: df_consultas, que contiene información sobre las consultas médicas, y df_ubicaciones, que contiene información sobre la ubicación de las instituciones de salud. Se eliminan las columnas no necesarias como 'Unnamed: 0'.

### Filtrado de Datos

Una vez que los datos se han cargado correctamente, el siguiente paso es filtrar los registros que no son relevantes para el análisis.

```python
duplicados = [
    "Hospital de Vila Franca de Xira, EPE",
    "Hospital de Loures, EPE",
    "Centro Hospitalar do Oeste, EPE, EPE"
]
df_ubicaciones_filtrados = df_ubicaciones[~df_ubicaciones['instituicao'].isin(duplicados)]
df_consultas_filtrado = df_consultas[~df_consultas['institucion'].isin(duplicados)]
```

Aquí se eliminan las instituciones que aparecen duplicadas en los datos.

### Renombrar Instituciones

Para facilitar el análisis, los nombres largos de las instituciones se mapean a nombres más cortos y manejables.

```python
# Defino un diccionario para mapear los nombres largos a nombres más cortos y manejables
name_mapping = {
    'Centro Hospitalar de Setúbal, EPE': 'CH Setúbal',
    'Centro Hospitalar do Oeste, EPE': 'CH Oeste',
    'Hospital Garcia de Orta, EPE': 'H Garcia Orta',
    'Centro Hospitalar Universitário Lisboa Central, EPE': 'CHU Lisboa Central',
    'Hospital Distrital de Santarém, EPE': 'H Santarém',
    'Hospital Professor Doutor Fernando Fonseca, EPE': 'H Fernando Fonseca',
    'Instituto Gama Pinto': 'I Gama Pinto',
    'Centro Hospitalar Universitário de Lisboa Norte, EPE': 'CHU Lisboa Norte',
    'Hospital de Cascais, PPP': 'H Cascais',
    'Instituto Português Oncologia de Lisboa, EPE': 'IPO Lisboa',
    'Centro Hospitalar de Lisboa Ocidental, EPE': 'CH Lisboa Ocidental',
    'Centro Hospitalar Psiquiátrico de Lisboa': 'CHP Lisboa',
    'Centro Hospitalar Barreiro/Montijo, EPE': 'CH Barreiro/Montijo',
    'Centro Hospitalar Médio Tejo, EPE': 'CH Médio Tejo',
    'Hospital de Vila Franca de Xira, PPP': 'H Vila Franca',
    'Hospital de Loures, PPP': 'H Loures'
}

df_consultas['institucion'] = df_consultas['institucion'].map(name_mapping)
```

Este diccionario name_mapping se utiliza para simplificar los nombres de las instituciones.

Para entender mejor la distribución de las consultas médicas en Lisboa, se segmentan los datos por año y se calculan varias estadísticas descriptivas.



```python
lisboa_2018 = df_consultas.loc[df_consultas['label_year'] == 2018]
lisboa_2019 = df_consultas.loc[df_consultas['label_year'] == 2019]
lisboa_2020 = df_consultas.loc[df_consultas['label_year'] == 2020]
lisboa_2021 = df_consultas.loc[df_consultas['label_year'] == 2021]
lisboa_2022 = df_consultas.loc[df_consultas['label_year'] == 2022]

estadisticas_descriptivas = {
    'primeras_consultas': ['mean', 'median', 'sum', 'min', 'max', 'std'],
    'consultas_seguimiento': ['mean', 'median', 'sum', 'min', 'max', 'std'],
    'total_mes': 'sum'
}

# Aplicamos el groupby y agg a cada DataFrame filtrado por año
lisboa_2018_stats = lisboa_2018.groupby('institucion').agg(estadisticas_descriptivas).round(2)
lisboa_2019_stats = lisboa_2019.groupby('institucion').agg(estadisticas_descriptivas).round(2)
lisboa_2020_stats = lisboa_2020.groupby('institucion').agg(estadisticas_descriptivas).round(2)
lisboa_2021_stats = lisboa_2021.groupby('institucion').agg(estadisticas_descriptivas).round(2)
lisboa_2022_stats = lisboa_2022.groupby('institucion').agg(estadisticas_descriptivas).round(2)

lisboa_2018_stats.reset_index(inplace=True)
lisboa_2019_stats.reset_index(inplace=True)
lisboa_2020_stats.reset_index(inplace=True)
lisboa_2021_stats.reset_index(inplace=True)
lisboa_2022_stats.reset_index(inplace=True)
```

Se agrupan los datos por institución y se calculan estadísticas descriptivas como la media, mediana, suma, mínimo, máximo y desviación estándar para las variables primeras_consultas y consultas_seguimiento.

#### Relación entre Primeras Consultas y Consultas de Seguimiento

```python
# Almaceno mis DataFrames en un diccionario
dfs = {
    'lisboa_2018': lisboa_2018,
    'lisboa_2019': lisboa_2019,
    'lisboa_2020': lisboa_2020,
    'lisboa_2021': lisboa_2021,
    'lisboa_2022': lisboa_2022,
}

# Definir una paleta de colores
paleta_colores = {
    'lisboa_2018': 'blue',
    'lisboa_2019': 'green',
    'lisboa_2020': 'red',
    'lisboa_2021': 'purple',
    'lisboa_2022': 'orange'
}

# Itera sobre el diccionario
for nombre, df in dfs.items():
    x = df['primeras_consultas']
    y = df['consultas_seguimiento']
    
    # Crear una figura y un eje
    fig, ax = plt.subplots()
    
    # Usar Seaborn para trazar los puntos de 'primeras_consultas'
    sns.scatterplot(x=x, y=y, color=paleta_colores[nombre], ax=ax)
    
    # Usar Seaborn para trazar la línea de regresión
    sns.regplot(x=x, y=y, scatter=False, color='gray', ax=ax)
    
    # Establecer el título del gráfico basado en el nombre del DataFrame
    plt.title(f'Relación entre Primeras Consultas y Consultas de Seguimiento para {nombre.split("_")[1]}')
    
    # Mostrar el gráfico
    plt.show()
```

Estos gráficos muestran la relación entre las primeras consultas y las consultas de seguimiento para cada año, lo que puede ayudar a entender mejor cómo estas dos variables interactúan entre sí.
